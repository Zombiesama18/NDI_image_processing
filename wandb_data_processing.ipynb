```python
mean_max = []
name = []
for item in temp_dfs:
    mean_value = item.mean(axis=1)[30:]
    temp_mean_max = [(mean_value.mean(), mean_value.max())]
    mean_max.append(temp_mean_max)
    name.append(item.columns[0])
    
names = name
```

```python
mean_max = []
for df_10, df_20, df_30 in dfs:
    mean_value = df_10.mean(axis=1)[30:]
    temp_mean_max = [(mean_value.mean(), mean_value.max())]
    mean_value = df_20.mean(axis=1)[30:]
    temp_mean_max.append((mean_value.mean(), mean_value.max()))
    mean_value = df_30.mean(axis=1)[30:]
    temp_mean_max.append((mean_value.mean(), mean_value.max()))
    mean_max.append(temp_mean_max)

    
```
